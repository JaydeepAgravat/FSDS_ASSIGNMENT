## 17\. How do you distinguish between copy.copy() and copy.deepcopy()?


**ANS:**
- `copy.copy()` function will do a shallow copy of a list.
- `copy.deepcopy()` function will do a deep copy of a list.

```python
import copy

old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]
new_list = copy.copy(old_list)

old_list[1][1] = 'AA'
# both lists share the reference of same nested objects.
print("Old list:", old_list)
print("New list:", new_list)
```

```python
import copy

old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]
new_list = copy.deepcopy(old_list)

#  The deep copy creates independent copy of original object and all its nested objects.
old_list[1][1] = 'AA'

print("Old list:", old_list)
print("New list:", new_list)
```
