## 6\. How do you tell the difference between the equal to and assignment operators?Describe a condition and when you would use one ?

**ANS:**
- Equal to `(==)` operator, compare two value and return boolean value.
- It returns `True` if both the values are equal, otherwise returns `False`. 
- Assignment `(=)` operator, it is used to assign the value to operand.
- to compare the value we use `(==)` and to assign the value to operand we use `(=)`.

````python
# expression(2*5) is evaluated to 10 and 
# assignment operator (=) assign the 10 to x variable
x = 2 * 5

# compare x to 10
if x == 10:
  print("x is 10.")
else:
  print("x is not 10.")
````

## 8\. Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints Greetings! if anything else is stored in spam.

**ANS:**

````python
def spam_message(spam):
  if spam == 1:
    print("Hello")
  elif spam == 2:
    print("Howdy")
  else:
    print("Greetings")

spam_message(1)
spam_message(2)
spam_message(3)
````

## 12\. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop ?

**ANS:**

````python
# for loop
for num in range(1,11):
  print(num,end=" ")

print()

# while loop
num = 1
while num < 11:
  print(num,end=" ")
  num += 1
````
